Step1: Go to the encodeproject.org to download the epigenomic events data you are interested.

For example, when focusing on the transcription factors, you may check the following selections.

Select the transcription factors, focus on the homo sapiens, using the genome assembly hg38 and focus on the peak calling results files.

Step2: Using the code provided to process the files.

```python
from utils import *

bin_length = 200

working_directory_path = '../data'

#Prepare the bin index for the whole genome.
prepare_whole_genome_bed_file(chr_index, chr_length_hg38, bin_length, working_directory_path)
#Convert the peak calling results into binary, whether a bin containing the given epigenomic event.
convert_peak_to_binary(working_directory_path, working_directory_path, chr_length_hg38, bin_length)
#Merge all epigenomic events matrix into a m*n matrix, with m different epigenomic events and n bins (bins has at least one epigenomic events)
merge_binary_label(working_directory_path, working_directory_path)
#Prepare the bed file containing all selected bins for the reference.
write_selected_region_to_bed(working_directory_path, working_directory_path, 'TF_summary.bed')
```

After that, you will get a merged matrix, m*n, containing m epigenomics and n regions. With a bed file containing where those regions.
